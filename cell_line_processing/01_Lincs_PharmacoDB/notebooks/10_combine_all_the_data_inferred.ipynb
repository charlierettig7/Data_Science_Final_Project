```python
def splitting_by_cell_line(df, 
                           signatures='sig_id', 
                           cell_line='cellosaurus_id', 
                           ratio=0.8,
                           seed=42):
    
    cell_lines = df['cellosaurus_id'].unique()
    train_size = round(ratio * len(cell_lines))
    val_size = round((1 - ratio)/2 * len(cell_lines))
    np.random.seed(seed)
    cell_lines_perm = np.random.permutation(cell_lines)
    train_cell_lines = cell_lines_perm[:train_size]
    val_cell_lines = cell_lines_perm[train_size:train_size + val_size]
    test_cell_lines = cell_lines_perm[train_size + val_size:]
    df_train = df[df['cellosaurus_id'].isin(train_cell_lines)].sample(frac=1).reset_index(drop=True)
    df_val = df[df['cellosaurus_id'].isin(val_cell_lines)].sample(frac=1).reset_index(drop=True)
    df_test = df[df['cellosaurus_id'].isin(test_cell_lines)].sample(frac=1).reset_index(drop=True)
    return df_train, df_val, df_test
```

```python
df_train, df_val, df_test = splitting_by_cell_line(df_sig_resp_final)
```
