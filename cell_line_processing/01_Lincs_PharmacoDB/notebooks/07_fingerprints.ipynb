```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
import pandas as pd
import numpy as np
```

```python
pharm_comp = pd.read_parquet('../../NO_BACKUP/pharmacodb/compounds_pharmdb.parquet')
```

```python
lincs_comp = pd.read_parquet('../../NO_BACKUP/lincs_2020/comp_info.parquet')
```

```python
lincs_comp_dropna = lincs_comp[~lincs_comp['inchi_key'].isna()]
```

```python
comp_overlap = lincs_comp_dropna[lincs_comp_dropna['inchi_key'].isin(pharm_comp['inchikey'].unique())]
```

```python
comp_overlap = comp_overlap[~comp_overlap['pert_id'].duplicated()]
```

```python
comp_overlap = comp_overlap.reset_index(drop=True)
```

```python
pharm_comp[['inchikey', 'pubchem']]
```

```python
comp_overlap.merge(pharm_comp[['inchikey', 'pubchem']], left_on='inchi_key', right_on='')
```

```python
comp_overlap['canonical_smiles']
```

```python
from rdkit.Chem import rdFingerprintGenerator
fp_diameter = 4
fp_radius = fp_diameter / 2
rdkgen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048)
fps = []
cfps = []

for smile in comp_overlap['canonical_smiles'].values:
    if not smile is None:
        mol = Chem.MolFromSmiles(smile)
        fp = rdkgen.GetFingerprintAsNumPy(mol)
        cfp = rdkgen.GetCountFingerprintAsNumPy(mol)
        fps.append(fp)
        cfps.append(cfp)
    else:
        fps.append(None)
        cfps.append(None)
```

```python
df_fps = pd.DataFrame(np.array(fps))
```

```python
rename_fps_col = {}
for i in range(2048):
    rename_fps_col[i] = 'fps_' + str(i)
```

```python
df_fps = df_fps.rename(columns=rename_fps_col)
```

```python
df_fps['pert_id'] = comp_overlap['pert_id']
```

```python
df_fps['cmap_name'] = comp_overlap['cmap_name']
```

```python
df_fps['smiles'] = comp_overlap['canonical_smiles']
```

```python
df_cfps = pd.DataFrame(np.array(cfps))
```

```python
rename_cfps_col = {}
for i in range(2048):
    rename_cfps_col[i] = 'cfps_' + str(i)
df_cfps = df_cfps.rename(columns=rename_cfps_col)
```

```python
df_cfps
```

```python
df_cfps['pert_id'] = comp_overlap['pert_id']
```

```python
df_cfps['cmap_name'] = comp_overlap['cmap_name']
```

```python
df_cfps['smiles'] = comp_overlap['canonical_smiles']
```

```python
df_fps.to_parquet('../NO_BACKUP/pharmacodb/comp_overlap_fps.parquet', index=False)
df_cfps.to_parquet('../NO_BACKUP/pharmacodb/comp_overlap_cfps.parquet', index=False)
```
