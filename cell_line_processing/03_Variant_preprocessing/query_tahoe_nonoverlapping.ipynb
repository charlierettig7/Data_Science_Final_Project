This file is the tahoe equivalent of the lincs processing notebook "merge_variants_into_lincs.ipynb". It uses cell lines existing in tahoe but not in lincs.

```python
import pandas as pd
```

```python
tahoe_path = '../data/final_tahoe_no_overlappings.parquet'
tahoe = pd.read_parquet(tahoe_path, engine='pyarrow')
```

```python
variant_path = '../data/mutations_cellosaurus_full.csv'
variants = pd.read_csv(variant_path)
variants = variants[variants['cellosaurus_ids.accession'].isin(tahoe['cell_line'])]
variants_subset = variants[[
    'cellosaurus_ids.accession',
    'HugoSymbol',
    'DNAChange',
    'VariantType'
]]
```

```python
variants_subset
```

```python
variants_subset['HugoSymbol'].unique().size
```

```python
variants_subset['cellosaurus_ids.accession'].unique().size
```

```python
# check how many unique mutations there are
all_mutations = list(set(variants_subset['DNAChange']))
print(f"number of unique mutations: {len(all_mutations)}")

# create a dict
# keys are the cell lines
# values are lists that contain all mutations in that cell line
variant_by_cell_line = {}
for cell_line, group in variants_subset[['cellosaurus_ids.accession', 'DNAChange']].groupby('cellosaurus_ids.accession'):
    # Convert group to list of dictionaries, excluding the grouping column
    variant_by_cell_line[cell_line] = group['DNAChange'].tolist()
```

```python
# join the mutation data to tahoe
tahoe['mutations'] = tahoe['cell_line'].map(variant_by_cell_line)
tahoe['mutations'] = tahoe['mutations'].apply(lambda x: x if isinstance(x, list) else [])
```

```python
print("Setup Done and starting creation of matrix", flush = True)
# we are creating an empty data frame
# rows = cell lines
# columns = mutations
# we use bools because its the data type needing the least memory
mutation_presence = pd.DataFrame(0, index=range(tahoe.shape[0]), columns=all_mutations, dtype=bool)

# we convert it to a numpy array and 
mutation_array = mutation_presence.values
# create a dict that maps column names to integer position
mutation_cols = {col: i for i, col in enumerate(mutation_presence.columns)}

# now we iterate trough the mutations
for idx, muts in enumerate(tahoe['mutations']):
    if (idx % 2000 == 0):
        print(f"Iteration {idx} ....", flush = True)
    if isinstance(muts, list) and len(muts) > 0:
        # get the column indices for all present mutations, so we can set them at the same time
        # this is a huge speed improvement over basic iterating
        col_indices = [mutation_cols[mut] for mut in muts if mut in mutation_cols]
        mutation_array[idx, col_indices] = True

# Convert back to DataFrame for concatenating
mutation_presence = pd.DataFrame(mutation_array, 
                                index=mutation_presence.index, 
                                columns=mutation_presence.columns,
                                dtype=bool)
                                
print("Concatenating ...", flush = True)
tahoe_with_mutations = pd.concat([tahoe, mutation_presence], axis=1)

print("Write to file ...", flush = True)
tahoe_with_mutations.to_parquet(path='data/tahoe_no_overlapping_variant_matrix.parquet', engine='pyarrow')
```

# Aggregating mutations by Gene

```python
# this is just here because during EDA I was looking at a lot of different things
variants_subset = variants[[
    'cellosaurus_ids.accession',
    'HugoSymbol'
    ]]

# check how many unique genes there are
all_genes = list(set(variants_subset['HugoSymbol']))
print(f"number of unique genes: {len(all_genes)}")

# create a dict
# keys are the cell lines
# values are lists that contain all mutations in that cell line
grouped = variants_subset.groupby(['cellosaurus_ids.accession', 'HugoSymbol']).size().reset_index(name='count')
result = grouped.groupby('cellosaurus_ids.accession').apply(
    lambda x: dict(zip(x['HugoSymbol'], x['count']))
).to_dict()

print(result)
```

```python
mut_df = pd.DataFrame.from_dict(result, orient='index').fillna(0).astype(int)
mut_df['cellosaurus_ids.accession'] = mut_df.index
```

```python
mut_df
```

```python
tahoe_gene_level_muts = pd.merge(tahoe, mut_df, left_on='cell_line', right_on='cellosaurus_ids.accession', how='left')
```

```python
tahoe_gene_level_muts.to_parquet(path='../data/tahoe_no_overlapping_gene_level_matrix.parquet', engine='pyarrow')
```
